# Creación y entrenamientos de los modelos

```python
# separacion de la data en tartget y variables
target = data["Class"]
variables = data.drop(["Class"], axis=1, inplace=False)
```

```python
##Modelo 1(RN) Redes neuronales

from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import RandomizedSearchCV,cross_validate
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler


#modelo
m1MPLC = Pipeline([("std", StandardScaler()), ("mlp", MLPClassifier())])
parameters = {"mlp__learning_rate_init": [0.01, 0.001, 0.0001],
              "mlp__alpha": [0.01, 0.001, 0.0001,]}
m1RSMLPC = RandomizedSearchCV(m1MPLC, parameters,scoring ="roc_auc")

#entrenamiento del modelo 
m1RSMLPC.fit(variables, target)
m1RSMLPC.cv_results_

#metricas 
m1SCORE = cross_validate(m1RSMLPC, variables, target, scoring = ("recall","precision","f1","accuracy","roc_auc"))
m1SCORE
```

```python
#resultados de las metricas

import numpy as np
m1recall= np.mean(m1SCORE['test_recall'])
m1precision= np.mean(m1SCORE['test_precision'])
m1f1= np.mean(m1SCORE['test_f1'])
m1accuaracy= np.mean(m1SCORE['test_accuracy'])
m1rocauc= np.mean(m1SCORE['test_roc_auc'])
print(m1recall)
print(m1precision)
print(m1f1)
print(m1accuaracy)
print(m1rocauc)
```

```python
## Modelo 2(BG) Bagging 

from imblearn.ensemble import BalancedBaggingClassifier 
from sklearn.metrics import recall_score , precision_score, f1_score, confusion_matrix, accuracy_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_validate

#modelo
m2bbc = Pipeline([("std", StandardScaler()), ("bg", BalancedBaggingClassifier())])

#entrenamiento del modelo
m2bbc.fit(variables, target)

#metricas 
m2SCORE = cross_validate(m2bbc, variables, target, scoring = ("recall","precision","f1","accuracy","roc_auc"))
m2SCORE
```

```python
#resultados de las metricas

import numpy as np
m2recall= np.mean(m2SCORE['test_recall'])
m2precision= np.mean(m2SCORE['test_precision'])
m2f1= np.mean(m2SCORE['test_f1'])
m2accuaracy= np.mean(m2SCORE['test_accuracy'])
m2rocauc= np.mean(m2SCORE['test_roc_auc'])
print(m2recall)
print(m2precision)
print(m2f1)
print(m2accuaracy)
print(m2rocauc)
```
